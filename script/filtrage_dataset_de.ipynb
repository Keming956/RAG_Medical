```python
from datasets import load_dataset

dataset = load_dataset("wikipedia", language="de", date="20220301")
```

```python
dataset
```

```python
import re

mental_health_keywords_de = [
    # Depression & Angststörungen
    "depression", "major depression", "klinische depression",
    "angst", "generalisiertes angstsyndrom", "panikstörung",

    # Bipolare Störung & Manie
    "bipolare störung", "manie", "manische episode",

    # Schizophrenie, Paranoia, Wahn
    "schizophrenie", "paranoia", "wahnvorstellung", "wahn",

    # PTSD & Trauma
    "ptbs", "posttraumatische belastungsstörung", "trauma",

    # Zwangsstörung (OCD)
    "zwangsstörung", "obsessiv-kompulsive störung", "ocd",

    # ADHS
    "adhs", "aufmerksamkeitsdefizit-hyperaktivitätsstörung",

    # Autismus
    "autismus", "autismus-spektrum-störung", "asperger-syndrom",

    # Essstörungen
    "essstörung", "anorexie", "bulimie", "binge eating",

    # Persönlichkeitsstörungen
    "persönlichkeitsstörung", "borderline-persönlichkeitsstörung", "narzisstische persönlichkeitsstörung",

    # Psychose & Halluzinationen
    "psychose", "psychotische episode", "halluzination",

    # Suizid & Selbstverletzung
    "suizid", "suizidgedanken", "selbstverletzung", "suizidale ideation",

    # Kindheit, Jugend & Entwicklung
    "jugenddepression", "kindheitstrauma", "entwicklungsstörung",

    # Schlaf & Phobien
    "panikanfall", "insomnie", "schlafstörung", "soziale angst", "phobie",

    # Allgemeine Begriffe
    "psychische erkrankung", "psychische störung", "psychiatrie", "psychotherapie",
    "kognitive verhaltenstherapie", "cbt", "antidepressivum", "medikation", "diagnose",

    # Fachbegriffe & Klassifikationen
    "dsm-5", "klinische psychologie", "emotionale dysregulation", "impulsivität"
]


keyword_pattern = re.compile(r"|".join([re.escape(kw.lower()) for kw in mental_health_keywords_de]))

filtered_dataset = dataset.filter(
    lambda example: bool(keyword_pattern.search(example["title"].lower()))
)
```

```python
for split in filtered_dataset.keys():
    filtered_dataset[split].to_json(f"corpus_de_{split}.jsonl", lines=True, force_ascii=False)
```
