```python
from datasets import load_dataset

dataset = load_dataset("wikipedia", language="en", date="20220301")
```

```python
keywords = [
    "depression", "major depressive disorder", "clinical depression",
    "anxiety", "generalized anxiety disorder", "panic disorder",
    "bipolar disorder", "mania", "manic episode",
    "schizophrenia", "paranoia", "delusion",
    "ptsd", "post-traumatic stress disorder",
    "ocd", "obsessive compulsive disorder",
    "adhd", "attention deficit hyperactivity disorder",
    "autism", "autism spectrum disorder", "asperger syndrome",
    "eating disorder", "anorexia", "bulimia", "binge eating",
    "personality disorder", "borderline personality disorder", "narcissistic personality disorder",
    "psychosis", "psychotic episode", "hallucination",

    "suicide", "suicidal thoughts", "self-harm", "suicidal ideation",

    "teen depression", "childhood trauma", "developmental disorder",

    "panic attack", "insomnia", "sleep disorder", "social anxiety", "phobia", "trauma",

    "mental illness", "mental disorder", "psychiatry", "psychotherapy",
    "cognitive behavioral therapy", "cbt", "antidepressant", "medication", "diagnosis",

    "DSM-5", "clinical psychology", "emotional dysregulation", "impulsivity"
]
```

```python
import re

keyword_pattern = re.compile(r"\b(" + "|".join([re.escape(kw.lower()) for kw in keywords]) + r")\b")

filtered_dataset = dataset["train"].filter(
    lambda example: bool(keyword_pattern.search(example["title"].lower())))
```

```python
filtered_dataset.to_json("psych_wikipedia_en.jsonl")
```
